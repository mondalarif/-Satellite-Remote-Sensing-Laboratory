```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
path='/home/arif/Desktop/Python Lab'
os.chdir(path)

df=pd.read_csv('dune_params.csv')
```

```python
df
```

```python
c_length=df["Crest Length"]
d_length=df["Dune Length"]
d_height=df["Dune Height"]
```

```python
fig, ax = plt.subplots(1,2, figsize=(12, 6))

ax[0].scatter(d_length, c_length)
ax[0].set_xlabel("Dune Length (m)")
ax[0].set_ylabel("Crest Length (m)")
ax[0].set_title("Crest Length vs Dune Length")

ax[1].scatter(d_length, d_height)
ax[1].set_xlabel("Dune Length (m)")
ax[1].set_ylabel("Dune Height (m)")
ax[1].set_title("Dune Height vs Dune Length")

plt.show()
```
